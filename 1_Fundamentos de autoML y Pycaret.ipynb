# AutoML

## Proceso de Data Science 

El proceso de generación de un modelo de ML incopora numerosos y complejos pasos. Este proceso es complicado y requiere de diferentes tipos de habilidades:
1.	La ingeniería de atributos
2.	La optimización de los hiperparámetros requiere una comprensión profunda de los algoritmos y los conceptos básicos de ML
3.	Se necesitan habilidades de ingeniería de software para que el código de salida sea fácil de comprender e implementar.


### Qué hace autoML

AutoML trata de encontrar convinaciones de modelos y de hisperparametros que maximicen la precision de los modelos.Para ello:

- AutoML elige una estrategia para preprocesar los datos: cómo tratar los datos desequilibrados, cómo llenar los valores faltantes, eliminar, reemplazar o mantener valores atípicos, cómo codificar categorías y columnas de categorías múltiples, cómo prevenir errores de memoria...
- AutoML genera nuevas variables y selecciona las significativas
- AutoML selecciona el modelo (modelos lineales, K-vecinos más cercanos, aumento de gradiente, redes neuronales...)
- AutoML ajusta los hiperparámetros del modelo elegido 
- AutoML crea un conjunto estable de modelos para aumentar la puntuación si es posible


## Pycaret

PyCaret es una herramienta de automatización del flujo de trabajo para el aprendizaje automático supervisado y no supervisado.

Todos los módulos en PyCaret admiten la preparación de datos (más de 25 técnicas de preprocesamiento esenciales, ajuste automático de hiperparámetros, análisis e interpretación de modelos, selección automática de modelos, registro de experimentos y opciones de implementación en la nube)


# Comandos básicos de Pycaret

Pycaret contiene numerosas funciones y atributos. Las funciones que se utilizan en la mayoria de casos son las siguientes:

1. <code>compare_models </code> entrena todos los modelos de la biblioteca de modelos utilizando los hiperparámetros por defecto y evalúa el rendimiento mediante validación cruzada. Entre las métricas utilizadas para la clasificación: Accuracy, AUC, Recall, F1, Kappa, MCC. Métricas utilizadas para la regresión: MAE, MSE, RMSE, R2, RMSLE, MAPE.

2. <code> create_model </code>entrena un modelo utilizando hiperparámetros predeterminados y evalúa métricas de rendimiento mediante validación cruzada.

3. <code>tune_model </code>ajusta los hiperparámetros del modelo. Utiliza una búsqueda de random grid search.

4. <code>ensemble_model </code> Se utiliza para crear un conjunto de modelos entrenados para mejorar el resultado. La función devuelve una tabla con estimaciones con validación cruzada y un objeto de un modelo entrenado.

5. <code>predict_ </code>se utiliza para inferencia / predicción.

6. <code>plot_model </code> se utiliza para evaluar el rendimiento del modelo de aprendizaje automático entrenado.


**Registro de experimentos**: PyCaret incorpora el componente de seguimiento de MLflow como una API y una interfaz de usuario de backend para registrar parámetros, versiones de código, métricas y archivos de salida al ejecutar su código de aprendizaje automático y para visualizar posteriormente los resultados.
